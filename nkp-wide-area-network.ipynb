###### 1. Early Internet and WAN Service Types
***

In this presentation, we'll take a look at some of the methods by which you can establish Internet or WAN connectivity. Or perhaps more generally stated, just connections that can span large distances over provider subscribed networks. 

Now, we'll begin with some of the early methods, but we'll also see some later ones in our next presentation. But that's not really to suggest that some of these early methods aren't still in use, some of them are. And some of them have progressed quite a ways. But they were just introduced much earlier than some of the ones that we'll see later. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Dial-up - Public-Switched Telephone Network (PSTN)
***

So it really all began with dial-up. Historically, this was the main way to access the Internet from a small office or home office, or Small Office/Home Office (SOHO) environment. Or of course, from residential environments as well. So it used your public-switched telephone network to initiate a phone call to your Internet service provider. And that provided you with connectivity to the Internet. Now this had significant drawbacks, particularly by today's standards. 

Modems, to this day, can only run up to 56 kilobits per second once they make an analog call per connection. Meaning that back in the old dial-up days, once you established this connection to get on the Internet, you could not make a phone call, your phone line was already in use. So unless you wanted a completely separate phone line dedicated to Internet access, you could not use the phone at the same time. And it was exceedingly slow by today's standards. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Integrated Services Digital Network (ISDN)
***

Now, this was superseded by ISDN, or Integrated Services Digital Network. And this was the first iteration, if you will, of high-speed Internet. It replaced dial-up Internet service, and it was still telephone based, but it uses a circuit switch or a dedicated line. In short, it means that your ISDN devices still use the public switched telephone network, but they could handle voice and data. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### ISDN - Basic Rate Interface (BRI) ISDN
***

So they were, for starters, faster and they essentially use two dedicated 64 Kbps lines for a total of 128 kilobits per second. This was known as Basic Rate Interface, or BRI ISDN. 

It was faster, and it could also handle your data and your voice at the same time. So in other words, you could use your Internet connection and make phone calls at the same time. So this was certainly a plus compared to dial-up. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### ISDN - Primary Rate Interface (PRI) ISDN
***

ISDN was also released as Primary Rate Interface, or PRI ISDN, and this version aggregated 23x64 Kbps channels together for voice or data use.

###### T Carriers
***

Now these digital carriers were destined to have a new format, however. As we saw the advent of the T carriers, T1 and T3 especially.

***
###### T1
***

Now T1 was the earlier, and it actually was the same wiring as PRI ISDN but with a different protocol running on top. 

It was used for both home and commercial use, but it was generally used more so in the larger business environment. But somebody who worked from home or had a business in their home might have invested in a T1. And it was also able to accommodate both voice and data transmissions, so you did not have to sacrifice your phone for the Internet connection. 

This was comprised of a series of digital channels, 24 in fact, that allowed for much faster speeds than standard Internet connections at the same time, approximately 1.5 megabits per second.

So again, significantly higher than ISDN.

###### T3
***

And then T3 came along, which was also known as Digital Signal 3 or DS3. And its transmission rates, again, were far higher than T1.

Ideal for large corporate networks, as they could accommodate much larger amounts of traffic. It was absolutely not designed for home or small office networks. They were significantly more expensive, they were up around 35, 36 megabits per second in terms of their speed, but they could costs thousands per month. So generally, you did not see T3 connections in a small office or home environment.

###### E1 and E3
***

Now E1 and E3 are fairly similar to T1 and T3. They're still WAN digital communications mediums, and as you can see, slightly different rates in terms of the megabit per second value. E1 around 2, and E3 around 34 megabits per second. 

And they had basic time-division multiplexing schemes for digital circuits. Effectively, they were implemented more frequently outside of North America. And sometimes, you would even hear people referring to the E as European because it was very commonly used there. But I don't believe it actually stood for that. 

But an E3 can carry up to 16 E1 circuits. So again, it was just a more robust implementation of the same type of methods used to establish the connections.

###### Optical Carrier (OC)
***

Finally, we have the OC standard, such as OC-3, which stands for Optical Carrier, and then the level indicated by the number. 

And we focus O3 because it was simply a very common implementation of Optical Carrier standards. But you can see, there's a number of levels, and the rate goes up accordingly with your level to very high rates by the time you get to OC-192. 

So this implemented standardized protocols which are used for transferring data over fiber optic networks. So it depended on the specifications in terms of the overall rate. But you can see, the level incrementing in a bit of a strange fashion. 

The OC level is really just a value that indicates your level. And you have an according STS value, or an electrical level, that actually stands for Synchronous Transport Signal. But in simpler terms, it indicates the basic unit of transmission. 

So with an STS-1, the basic unit of transmission of 1, your rate was 51.84 megabits per second. 

An STS-3 is three times higher. And you can see that the line rate triples accordingly. STS-9, now we have nine times the original. It's nine times faster, triple the previous one. 

Now, it didn't triple each time. You can see the next one is 12, then 18, then 24. But that number after the OC and/or the STS simply meant a multiplier of 51.8. So if you did the math for, let's say, STS-36, it would be 36 times 51.84, the basic level, the STS-1. 

So they got up fairly high by the time you get to STS-192. 

Now the STM is a framing unit, and this was defined by SDH, which is the Synchronous Digital Hierarchy. So they define these Synchronous Transport Modules, and each module could contain three basic units of transmission. So you can see that the STM-1 had three STS-1 units in it. So the value of the STM is always one-third of the STS.  So again, it was just a method of framing and putting those basic units of transmission together. 

Not unlike a packet, if you will. But the correct term in this case is framing. But again, the higher you go, the faster your connection. 

And, ultimately, by today's standards we're seeing the speeds that are typically used for connecting the backbone of the Internet, the largest network in the world, OC-192 with data speeds upwards of ten gigabits per second. But that certainly is not something that you would see in smaller offices or maybe even medium sized. These would be the largest, most robust connections on the go today. But again, this is a method started quite some time ago, in fact, as far back as the early 1980s. So it simply just evolved into a very robust method of WAN connectivity.

###### 4. Characteristics of Service - Switched
***

Now in this presentation, we'll take a look at some characteristics of WAN connectivity. And begin by separating out circuit-switched and packet-switched networking.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Circuit-switched Network
***

A circuit-switched network, allows a node to connect to the switching service and request a connection to another endpoint. 

Once connected through the switching media to that endpoint, those links and switching resources are devoted to maintaining that connection and no other. Once a circuit is made, no further pointers are necessary. 

The telephone network is a circuit switch network, which is why you don't have to keep reentering the phone number every time you want to say something. The circuit is already established. And why, if more people call an area. For example, after a natural disaster, than there are circuits available, the caller might get notified that all circuits are busy which means that all the possible links are already in use. [Circuit-switched networks are PTSN and ISDN. The connection dedicates point-to-point use of resources. ] 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Packet-switched Network
***

The next type of network is a packet-switched network. In this situation, the establishment of the connection of a node to a switch indicates the potential to reach a destination, but the destination is still unknown. 

This is more like coming to an airport, where you won't really go anywhere until you present your ticket at the gate, and then you can get on a plane to go somewhere. 

With a packet-switched network, each packet must have the address of the intended destination written on it. So that the switch can examine it, and forward it to a destination. If that sounds familiar, it's because it's exactly how ethernet switched network works. [Packet-switched networks are Ethernet, ATM, frame relay, and MPLS. Each frame is checked and forwarded using addresses. ] 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Network Type - Frame Relay
***

Now, the first specific network type we'll look at is Frame Relay. 

Which is a layer 2 WAN technology that implements packet-switching telecommunication. 

Now, before going forward we call that packet-switching in of itself it's based on two endpoints that need to communicate with each other. But before they can do so, they have to establish a connection. But they only need the connection for the duration of the communication. Then, it shuts down. 

Again, very common day-to-day example, a phone call. The physical phone network is always there. But in order for me to use it, I have to pick up the phone, dial the right number, the person must pick up the other end, then we can communicate. When we're done, we hang up. And of course, while we aren't using it, then the infrastructure and the bandwidth is there available for other people to use. So we'll all share it, of course. So this is still packet-switching very much like a LAN environment. But it's over WAN communication methods. So in a lot of cases, it's like a public switch telephone network. 

So we can't just do ethernet switching over things like a phone network. But this was a cost effective solution because of the fact that the data transmission only occurs on that intermittent basis between the WAN endpoints again as needed. Just like when I hang up the phone and I'm done with my conversation, it is available for other people to use. So that's as needed basis. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Addressing - Frame Relay - Data Link Connection Identifier (DLCI)
***

Now in terms of addressing, again, this is another reason why you can't just implement LAN communications. It uses a method known as DLCI or data link connection Identifier. 

And again, for very general day to day analogies, this would be the phone number. This addresses the other endpoint. So once you've located the other endpoint, the two sides are connected within a network of virtual circuits. Virtual circuits are also known as VCs. 

And again, the virtual circuit means the physical media is always there, but I'm not using it all the time, I only use it when I need it. So it's very much like a phone call, but of course, we're exchanging data between computers. So it's something that was quite popular when it was first released, but is being certainly discontinued, because it's very slow by current standards. It only offered speed of up to 1.544 megabits per second. So again, very slow by today's standards, but back when it was first released that was pretty good. 

And again, it was based on the fact that we only need that bandwidth for a little while, then we shut down and other people can use it. Because again, even with your home phone system you aren't using it most of the time. So you could take advantage of the fact that most people weren't using it most of the time, so you could get better bandwidth than, of course, the dial up. But again, by today standards, it's fairly slow. 

< [Table of Contents](#top) | [References](#references) >

***
###### Network Type - Asynchronous Transfer Mode (ATM)
***

So ATM is another option. ATM stands for Asynchronous Transfer Mode. 

And this is also a layer 2 WAN technology that uses a virtual path identifier, virtual circuit identifier or VPI, VCI for addressing. So again, it was still based on the fact that, you have to identify the endpoint and you still have to make a connection to that endpoint. It use different methods of transmitting. Now this is actually older than frame relay, but is still used to a certain degree in high end ISPs. Because it was actually much faster than frame relay, despite the fact that it's actually older. 

You could typically find bandwidths of 155 megabits per second, to 622. So significantly faster than frame relay. And the basic idea behind ATM was to not just be a data switching technology but rather all types of information. Data, voice and video so that we could consolidate all these types of transmissions into a single protocol, and it provided guaranteed bandwidth by using a fixed cell size. This is the reason why they call it a cell. Because a packet is officially variable in size. 

With ATM, it was a very small cell size of only 53 bytes. So it used that in combination with time division multiplexing, which is referred to as TDM for short. You can see the multiplexer in the graphic. It's used to establish virtual circuits and begin transmitting data. But that fixed cell size reduced a lot of overhead. Basically, it didn't care what kind of data it was. Whether or not it was voice or video or traditional data, it just packed 53 bytes into every cell and sent it on its way. So it was very efficient, hence some of the greater speeds. But it also required dedicated equipment that could be fairly costly. And while still being implemented by some ISPs, never really caught on in private corporations because of the costs involved and the complexity of the implementation.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Network Type - Multiprotocol Label Switching (MPLS)
***

Finally, MPLS or Multiprotocol Label Switching. 

Is a method to enable efficient routing based on labels, hence the L of the MPLS. 

Now, what you see here is the interior of the cloud is what's called the label switch router. 

And that's just a router enabled for this protocol. 

Then the ELSR is the Edge Label Switch Router. And then the CPE stands for Customer Premises Equipment, so those would be our routers. [The graphic shows a router process between Site A and Site B using the Service provider's MPLS cloud. From Site A, the CPE connects to the ELSR, which is connected to the LSR. This in turn is connected to another ELSR, this connects to a CPE and on to Site B. The routing path is multi-directional.] 

The ELSRs would be the Internet Service Provider to which you're connected. The LSRs just represent the internet.

But the idea behind the routing methodology here is to simply attach a label to the packets that indicate a known virtual path to this router. Traditional routers have to examine the packet, locate the network address, look that up in its routing table, and then forward that to the next router based on that information. And the next router has to do the same thing, and it just carries it on as it goes from router to router to router. 

In this case, it just reads the label, which, if you will, is the first thing it sees. And it sees a path to a distant router, it just says, here you go. You know, go about your business, go on to your way to that distance router and keep doing that until you arrive at the endpoint. But this enabled you to arrive at the endpoint when it came to routing and overall, just much simpler implementations. 

And because of the fact that it only looked at the label and then nothing else, you can actually route traffic from different protocols. Because it quite literally did not care. It saw the label and said, off you go. So it was actually able to incorporate different types of traffic. Such as frame relay, ATM, DSL, T1, it didn't care. Whatever it was, any type of protocol could be implemented and could be pass through in LSR type of router. So this is a bit of a newer kit on the block if you will. But not overly new by today's standards. It was early released in the late 90s. But it's still being adopted because of the some of the older methods still being in place. But I think you'll start to see more MPLS coming into the picture as things move forward. And ultimately, these are all just different methods for moving traffic around in terms of WAN interconnectivity. And if you're talking about creating these types of connections between your own sites, within your own organization, then you certainly might consider some of these. But things like VPNs are certainly still an option as well. And our point-to-point connections which we'll actually look at in our upcoming presentation.

***
###### 5. Characteristics of Service - Point-to-Point

###### Point-to-Point Protocol (PPP)
***

Now, to again use an analogy, if we consider the phone network, a plain old phone call is in fact a point-to-point connection.

I have to pick up the phone, I have to dial the number. The other person has to answer the phone. That establishes the connection and then we can talk. That is point to point. 

But with respect to connections that use point-to-point communications in networking, it is more so used to establish a permanent virtual connection as opposed to a switched virtual connection. 

And again, you can still kind of imagine the same thing with a phone call. I can call you on the phone, and as long as we're communicating, we need that connection available to us. But then when we are done, we typically hang up. But, in fact, we don't have to hang up, we could leave both phones off the hook. And if we feel like we are going to communicate again in a few minutes, or even hours later we would just have to pick up the phone and the connection would still be there. So that's typically where you see point-to-point connections implemented in network environments, when you need something that is a little more permanent. Maybe we aren't transferring data all the time, all day long, but it is fairly regular. So much so that it makes a little more sense to just leave it connected, than it does to open it, send, close, open, send, close, open, send, close. 

So, again, we see a little bit more of a permanent connection even though it is still officially a switched connection that we do still close once we're done. But, again, that sort of the general idea of point-to-point connections.

###### Point-to-Point Protocol over Ethernet (PPPoE)
***

So then, the Point-to-Point Protocol, as mentioned, can act as the foundation for different types of connections including Point-to-Point Protocol over Ethernet, or PPPoE. 

And this is a specification that's used to connect multiple users on an Ethernet LAN to a remote site, which in most cases is just the Internet. 

But it uses customer premises equipment. Now that equipment is typically supplied by the Internet service provider, but it stays on your premises. You then use it to simply share the connection. 

So, typically, what you would do in terms of gaining access to the Internet for a business environment is that you would just get one subscription. You would call up your Internet service provider and say, send some Internet to this location. But you only needed that one connection, but you needed to then share it amongst multiple users. So a set of credentials was set up to establish connectivity back to the Internet service provider. But you didn't need separate credentials for every single user within the Ethernet LAN environment. All of you accessed the single connection through the PPPoE server. So it effectively shared the connection and allowed us all to use that Internet connection. 

And, in essence, this is what you saw with DSL and cable modems. We could implement a way for all of us to use that connection, as opposed to just a single computer using it at any one given time. [In essence, PPPoE is used in office environments to share a common connection to the Internet.]

###### Dynamic Multipoint Virtual Private Network (DMVPN)

Now a DMVPN is a dynamic multipoint virtual private network, and this is still a VPN, which is a secure network for exchanging data between parties.

Then it might be able to connect me through to various resources within the environment. [A DMVPN does not require traffic to pass through a main VPN or router.] 

But in situations where you have multiple branches, rather than Branch 1 needing to communicate with Branch 2 through headquarters, it only contacts headquarters long enough to find out what is the connection information. And once that is known, then a direct connection is established between Branch 1 and Branch 2. So we only contact the hub for the correct information. Where do I go to get to Branch 2, Branch 3, Branch 4 and so on? Once I receive that, I establish these direct connections dynamically and traffic does not pass through the hub in headquarters. ###### Demarcation Point (demarc)
***

So the first consideration is what's known as a demarcation point. [The Demarcation point is also called the demarc.] 

Now this does not necessarily refer to any specific type of device or hardware, but rather just the physical point where a public network and a private network interconnect. So the telecommunications network would be the provider and your business would be the organization. So this is quite literally where the two meet. 

Now, the demarcation point is generally installed on site at the local business. It's typically a grey box that might be installed on the outside of the business. So their cabling runs into the box from your perspective and then your cabling runs out of the box into the building. 

Now, if it's a home or residential service then the demarcation point is typically installed inside maybe in the basement or in an attic or just somewhere out of the way usually but again, that physical point marks the boundary. 

So if a problem occurs, then determining who is responsible for that problem is quite literally, well which side of the demarcation point is encountering the problem. So to use very simple examples, if I trip over a cable and yank it out of the router and the cable breaks. Well clearly, that's my problem. 

But if something happens on the outside. Let's say a branch of a tree falls down and rips a cable out of the demarcation point on the outside, then that is actually their problem. 

So it really does mark that physical boundary so when problems occur you may call lap the provider for assistance. They can usually test to see whether or not the signal is making it to demarcation point. And, if it is then more or less they can say, well, it's not really our problem, there's something wrong on the inside. Now they might be able to assist you but they might also charge for that service. But as mentioned if they can not get signal back from the demarcation point, then they can say well yes, it looks like there's a problem with our service and accordingly they will have to look after that.
